```python
ageMean = clients.Age.mean()
clients.Age = clients.Age.fillna(ageMean)
clients.info()
```

```python
oldCategory  = [clients.Sex == 'Female', clients.Sex == 'Male']
newCategory     = ['0', '2']
clients.Sex = np.select(oldCategory, newCategory, default = 1)
clients.Sex.astype('category')
```

```python
orders = orders[orders.SummAfterPointsUsage > 0]
```

```python
order_y = orders.groupby('CustomerID').apply(
    lambda x: pd.Series(dict(Y=(x.RKDate.max() >= "2017-07-01") & (x.RKDate.max() <= '2017-12-31'))))
```

```python

order_fixed_date = orders[orders.RKDate <= "2017-07-01"]
```

```python
from datetime import datetime

def calc_results(groupOrdersByClient, date):

    theDate = datetime.strptime(date, '%Y-%m-%d')
    groupOrdersByClient['Date'] = pd.to_datetime(groupOrdersByClient.RKDate, format='%Y-%m-%d')
    recency = (theDate - groupOrdersByClient.Date.max()).days
    groupOrdersByClient['Day'] = groupOrdersByClient.Date.dt.day
    groupOrdersByClient['Month'] = groupOrdersByClient.Date.dt.to_period('M')
    frequency = groupOrdersByClient.Day.count()/groupOrdersByClient.Month.nunique()
    monetary_value = groupOrdersByClient.SummAfterPointsUsage.mean()
    meanDiscount = (groupOrdersByClient.SummBasic - groupOrdersByClient.SummAfterPointsUsage).mean()
    return pd.Series(dict(Recency = recency, Frequency = frequency, Monetary_Value = monetary_value, MeanDiscount = meanDiscount))
```

```python
order_fixed_date_results = order_fixed_date.groupby('CustomerID').apply(lambda x: calc_results(x, "2017-07-01"))
order_fixed_date_results.describe()
```

```python
df = pd.merge(left=order_y, right=order_fixed_date_results, left_on='CustomerID', right_on='CustomerID')
df
```

```python

df = pd.merge(left=clients, right=df, left_on='CustomerId', right_on='CustomerID')
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = df[['Recency', 'Frequency', 'Monetary_Value', 'MeanDiscount']]
y = df.Y

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import recall_score, precision_score

score = logreg.score(X_test, y_test)
precision = precision_score(y_pred, y_test)
recall = recall_score(y_pred, y_test)
print('Score: ' + str(score) + ' Precision: ' + str(precision) + ' Recall: ' + str(recall))
```
